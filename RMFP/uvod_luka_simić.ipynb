### __Zadatak 1__: Jedna kolonija od $N_0$ bakterija razmnožava se brzinom od $r = 0.8$ jedinki za $1$ sat. Koliko će biti bakterija u koloniji nakon $10$ sati? 

a) Definisati funkciju __bakterije__ koja modeluje zadati problem i omogućava rešavanje diferencijalne jednačine primenom integrate.odeint metode. Funkcija poziva parametar $r$. <font color='red'>[0.5]</font>

```python
def bakterije(N, t, r):
    return 0.8 * N
```

b) Napisati komande koje uvoze modul numpy, modul matplotlib.pyplot i funkciju za integraciju iz scipy modula.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
```

c) Definisati vrednost parametra $r = 0.8$ 1/h i početni uslov da je $N_0$ broj bakterija u trenutku $t=0$. Neka početni broj bakterija odgovara rednom broju tvog indeksa (bez godine).<br>
Definisati vremensku osu: 0 do 10 sati u 100 ekvidistantnih tačaka.

```python
N0 = 368
r = 0.8
t = np.linspace(0, 10, 100)
```

d) Napisati kod koji omogućava da se na istom grafiku, iscrta vremenski odziv $N(t)$ dobijen numeričkim i dobijen analitičkim rešavanjem diferencijalne jednačine koja modeluje dinamiku kolonije bakterija. Označiti ose grafika (“vreme [h]” i “broj bakterija”). Analitičko rešenje prikazati crvenom punom linijom, a numeričko rešenje isprekidanom plavom linijom <font color='red'>[1]</font>.<br>
_Da bi se prikazao grafik potrebno je izvršiti komandu plt.show()_

```python
numeričko = odeint(bakterije, N0, t, (r,))
analitičko = N0 * np.exp(0.8 * t)
plt.plot(t, numeričko, color='blue', linestyle='solid')
plt.plot(t, analitičko, color='red', linestyle='dashed')
plt.xlabel('broj bakterija')
plt.ylabel('vreme [h]')
plt.show()
```

e) Modifikovati funkciju __bakterije__ tako da uzme u obzir da nakon vremena $t_0$ počinje dejstvo hemikalije koje dovodi do odumiranja bakterija sa vremenskom konstantom $\tau$. Funkcija poziva parametre $r$,$t_0$,$\tau$ <font color='red'>[0.5]</font>.

```python
def bakterije2(N, t, r, t0, Tau):
    ret = r * N
    if t > t0:
        ret -= N/Tau
    return ret
```

f) Definisati vrednost parametra $\tau=2$ h i $t_0=8$ h, a zatim napisati kod koji omogućava da se na istom grafiku, uporede rezultati starog i novog modela. Označiti ose grafika (“vreme [h]” i “broj bakterija”). Rezultate simulacije dobijene na osnovu starog modela prikazati crvenom punom linijom, a rezultate dobijene primenom novog modela isprekidanom plavom linijom. Na graficima prikazati legendu koja se odnosi na odgovarajući tip modela.

```python
Tau = 2
t0 = 8
numeričko2 = odeint(bakterije2, N0, t, (r, t0, Tau))
plt.plot(t, numeričko, color='red', linestyle='solid')
plt.plot(t, numeričko2, color='blue', linestyle='dashed')
plt.show()
```

g) Ispitivanjem vrednosti za vremensku konstantu, odrediti potrebnu vrednost za koju broj bakterija u koloniji počinje da se smanjuje <font color='red'>[0.5]</font>. Ponoviti grafik iz tačke d) za izabranu vrednost parametra $\tau$

```python
Tau = 1
proba = odeint(bakterije2, N0, t, (r, t0, Tau))
plt.plot(t, proba, color='red', linestyle='solid')
plt.show()
```
